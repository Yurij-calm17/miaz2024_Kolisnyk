**"Збір даних для інформаційно-аналітичного забезпечення (ІАЗ) ОУВ"**.

### Загальне завдання:

#### Завдання:
1. **Збір даних із вебсайту**: Використовуючи Python і бібліотеку `BeautifulSoup`, зберіть заголовки новин із зазначеного вебсайту та збережіть їх у CSV-файл.
   
2. **Робота з формалізованими документами (Excel)**: Використовуючи Python та бібліотеку `pandas`, прочитайте дані з таблиці Excel, відфільтруйте їх за певною умовою (наприклад, виберіть записи за останній місяць) і збережіть відфільтровані дані у новий Excel-файл.

3. **Створення Google Форми**: Використовуючи Google Apps Script, автоматично створіть Google Форму для збору інформації (ім’я, дата, відділ). Після створення форми, збережіть посилання на форму.

4. **Створення XML-файлу**: Напишіть Python-скрипт для автоматичного створення XML-файлу, який міститиме інформацію про підрозділ (назва, кількість, стан).

---

### 6 варіантів особливостей завдання:

1. **Варіант 1**: Для завдання із веб-скрейпінгу додайте можливість фільтрувати заголовки новин за ключовими словами (наприклад, новини про технології або науку).
   
2. **Варіант 2**: Для збору даних із Excel використайте умовний фільтр (наприклад, виведіть лише записи, де значення більше ніж певна кількість).
   
3. **Варіант 3**: Додайте поле для збирання електронної пошти у Google Формі та зробіть його обов'язковим для заповнення.
   
4. **Варіант 4**: Створіть XML-файл для збору даних про ресурси (наприклад, транспортні засоби або обладнання) із зазначенням кількості та доступності.

5. **Варіант 5**: Виведіть зібрані дані з вебсайту в форматі JSON замість CSV.

6. **Варіант 6**: Додайте функцію автоматичної перевірки формату введеної електронної пошти у Google Формі.


### Порядок виконання завдання 
#### 1. **Створення папки проекту на GitHub**
   - **Створіть нову папку** в існуючому репозиторії GitHub. Назвіть її, наприклад, `data_collection_tasks`.
   - В папці має бути підпапка для кожного з завдань: 
     - `task_1_web_scraping`
     - `task_2_excel_processing`
     - `task_3_google_form`
     - `task_4_xml_creation`

#### 2. **Завдання 1: Збір даних із вебсайту**
   - Перейдіть у підпапку `task_1_web_scraping`.
   - Створіть Python-скрипт `web_scraper.py`, що використовує бібліотеку `BeautifulSoup` для збору заголовків новин із зазначеного вебсайту.
   - Виведіть зібрані дані у CSV-файл `news_titles.csv`.
   - Додайте файл README.md, в якому коротко опишіть призначення скрипта та як його запускати.

#### 3. **Завдання 2: Робота з Excel**
   - Перейдіть у підпапку `task_2_excel_processing`.
   - Створіть Python-скрипт `excel_processor.py`, що використовує `pandas` для читання Excel-файлу, фільтрації даних та збереження відфільтрованого результату.
   - Завантажте вихідний Excel-файл `data.xlsx` до цієї папки.
   - Виведіть відфільтровані дані у новий файл `filtered_data.xlsx`.
   - Додайте файл README.md з поясненням, як працює цей скрипт.

#### 4. **Завдання 3: Створення Google Форми**
   - Перейдіть у підпапку `task_3_google_form`.
   - Створіть Google Apps Script файл `google_form.gs`, який автоматично генерує Google Форму для збору даних (ім'я, дата, відділ).
   - У README.md надайте інструкції, як налаштувати цей скрипт і отримати URL згенерованої форми.

#### 5. **Завдання 4: Створення XML-файлу**
   - Перейдіть у підпапку `task_4_xml_creation`.
   - Створіть Python-скрипт `xml_creator.py`, що автоматично створює XML-файл з описом даних про підрозділ (назва, кількість, стан).
   - Збережіть згенерований XML-файл під назвою `unit_data.xml`.
   - В README.md опишіть процес запуску скрипта та структуру XML-файлу.

#### 6. **README.md для головної папки репозиторію**
   - Створіть основний `README.md` у кореневій папці репозиторію.
   - Включіть загальний опис проекту та покрокові інструкції для запуску кожного завдання.
   - Додайте посилання на підпапки з кожним завданням.

#### 7. **Останній етап: Завантаження на GitHub**
   - Після завершення створення скриптів і файлів, завантажте всі файли у відповідні підпапки вашого репозиторію.
   - Здійсніть локальний commit. 
   - Оновіть GitHub репозиторій (push).
