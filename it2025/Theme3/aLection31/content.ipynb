#### **4.1.2. Опис завдання**

- Необхідно виконати аналіз для оцінки бойової готовності підрозділів.

#### **4.1.3. Завдання**
1. **Визначення активної техніки у підрозділах**:
   ```sql
   SELECT u.name AS unit_name, COUNT(e.equipment_id) AS active_equipment
   FROM Units u
   JOIN Equipment e ON u.unit_id = e.unit_id
   WHERE e.status = 'Активна'
   GROUP BY u.name;
   ```
2. **Підрахунок виконаних маршрутів технікою**:
   ```sql
   SELECT e.name AS equipment_name, COUNT(r.route_id) AS total_routes
   FROM Equipment e
   LEFT JOIN Routes r ON e.equipment_id = r.equipment_id
   GROUP BY e.name;
   ```

3. **Аналіз середньої довжини маршруту для кожного типу техніки**:
   ```sql
   SELECT e.type, AVG(r.distance_km) AS avg_distance
   FROM Equipment e
   JOIN Routes r ON e.equipment_id = r.equipment_id
   GROUP BY e.type;
   ```
